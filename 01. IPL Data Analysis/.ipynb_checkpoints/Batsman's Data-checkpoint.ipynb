```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

balls = pd.read_csv("IPL Deliveries 2008-2020.csv")
matches = pd.read_csv("IPL Matches 2008-2020.csv")

display(balls.head(2),matches.head(2))
```

# Adding Required Data 

```python
balls["extras_type"].value_counts()

def wides_noballs(x):
    if x["extras_type"]=="wides" or  x["extras_type"]=="noballs":
        return 1
    else :
        return 0
    
balls["wides_or_noballs"] = balls.apply(lambda row: wides_noballs(row),axis=1)
balls
```

# Maximum Total Runs By Players

```python
max_runs = balls.groupby('batsman')['batsman_runs'].sum()
max_runs = max_runs.sort_values(ascending=False).iloc[0:50]

fig, ax = plt.subplots(figsize=(15, 5))

sns.barplot(max_runs.index, max_runs,ax=ax)
plt.title("Maximum Total Runs by Players")
plt.xlabel("Batsmans")
plt.ylabel("Total Runs")
plt.xticks(rotation = 90)
plt.figure(figsize=(10, 10))
```

# Highest Runs in a Match

```python
high_match_runs = balls.groupby(["id","batsman"], as_index=False)["batsman_runs"].sum()

high_match_runs = high_match_runs.sort_values(by="batsman_runs", ascending=False)

high_match_runs = high_match_runs.iloc[0:50]

# Show
fig, ax = plt.subplots(figsize=(15, 5))
plt.title("Highest runs in a match by individual")
plt.xlabel("Batsman")
plt.ylabel("Runs")
plt.xticks(rotation=75)
sns.barplot(high_match_runs["batsman"],high_match_runs["batsman_runs"])
```

# Maximum Strike Rates 

```python
match_runs = balls.groupby(["id","batsman"], as_index=False)["batsman_runs"].sum()
balls_faced = balls.groupby(["id","batsman"],as_index=False)["ball"].count()
extra_balls = balls.groupby(["id","batsman"],as_index=False)["wides_or_noballs"].sum()

strike_rate = pd.merge(match_runs,balls_faced,on=["id","batsman"])
strike_rate = pd.merge(strike_rate,extra_balls,on=["id","batsman"])
strike_rate["net_balls_faced"] = strike_rate["ball"] - strike_rate["wides_or_noballs"]
strike_rate["strike_rate"] = strike_rate["batsman_runs"]*100/(strike_rate["ball"]-strike_rate["wides_or_noballs"])

display(strike_rate.sort_values(by="batsman_runs", ascending=False))

avg_strike_rate = strike_rate.groupby(["batsman"]).mean()

display(avg_strike_rate.sort_values(by="strike_rate",ascending = False))

avg_strike_rate.loc[balls.batsman]
```
